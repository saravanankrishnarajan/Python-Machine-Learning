#### 1.Open and display the image “dog.jpeg”. Convert the image in to numpy array, so that i can be used in further processing.
[Hint: Use PIL module from python]

```python
import numpy as np
import pandas as pd


#import cv2
#img = cv2.imread('dogs.jpeg')
#print(type(imgarr))

from PIL import Image
img= Image.open('dogs.jpeg')
imgarr = np.asarray(img)
print(type(imgarr))
```

#### 2. Find out the dimensions of the image and convert it in to a two-dimensional array.

```python
imgarr.shape
```

#### 3.Use kmeans clustering with k set to 3 and cluster the image.
[Hint: Refer to k-means module of scikit learn]

```python
#new_arr=imgarr.ravel().reshape(imgarr.shape[0],imgarr.shape[1],imgarr.shape[2])
new_arr=imgarr.ravel().reshape(-1,imgarr.shape[2])
```

```python
new_arr.shape
```

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=3,init='k-means++')
kmeans.fit(new_arr)
```

```python
kmeans.labels_.shape
```

```python
newimg = new_arr.ravel().reshape(imgarr.shape)
```

#### 4. Predict the cluster label of every pixel in the image and plot it back as an image.

```python
print(kmeans.labels_.shape)
```

```python
#converting back the array to the same array shape as image we used.

newimg = new_arr.ravel().reshape(imgarr.shape)
print(newimg.shape)

# plotting the image using the new image array
import matplotlib.pyplot as plt

plt.imshow(newimg)
```
