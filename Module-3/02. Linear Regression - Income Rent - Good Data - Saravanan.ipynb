```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('income-rent-good.csv')
print(data.shape)
```

```python
data
```

```python
x=data[['Income']]
y=data[['Rent']]
model=linear_model.LinearRegression()
model.fit(x,y)
print("Coefficient (slope - m): ", model.coef_)
print("Intercept in Model (c)",model.intercept_)
```

```python
predictions = model.predict(x)
data['Predictions']= predictions
print(data)
```

```python
x = data['Income'].values
y = data['Rent'].values
p = data['Predictions'].values

plt.figure(figsize=(8,8))
plt.scatter(x,y,label='x,y scatter plot')
for i in range(len(x)):
    text = ''
    text = "("+str(x[i])+", "+str(y[i]) +")"
    plt.annotate(text,(x[i],y[i]+10))

plt.xlabel('Income')
plt.ylabel('Rent')

plt.plot(x,p,label='Regression Line')
plt.xticks(np.arange(1300,6500,100),rotation=90)
plt.yticks(np.arange(450,5200,100))
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import r2_score
print("R2:",r2_score(data[['Rent']],data[['Predictions']]))
```

```python
from sklearn import metrics
print("MSE: ", metrics.mean_squared_error(data[['Rent']], data[['Predictions']]))
print("RMSE: ", np.sqrt(metrics.mean_squared_error(data[['Rent']], data[['Predictions']])))
```
