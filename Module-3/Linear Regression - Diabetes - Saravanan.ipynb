```python
import pandas as pd
import math

df = pd.read_csv('data.csv')
df
# Fixing Np.NaN to mean values

#fixing invalid duration to valid
df.at[7,'Duration']=45

# fix missing date to current date
import datetime
currentDate = datetime.datetime.now()
date_time = currentDate.strftime('%d-%m-%Y')
print("date and time:", date_time)
df.at[27,'Date']=date_time

#filling NaN for Pulse
y= df['Pulse'].mean()
#format(ceil(x),'.1f')
print("Mean for Pulse is",math.ceil(y))
df['Pulse'].fillna(math.ceil(y),inplace=True)

#filling NaN for Calories
x= df['Calories'].mean()
#format(ceil(x),'.1f')
print("Mean for Calories is",math.ceil(x))
df['Calories'].fillna(math.ceil(x),inplace=True)

```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
model.fit(df[['Duration','Pulse','Maxpulse']], df.Calories)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
predictions = model.predict(df[['Duration','Pulse','Maxpulse']])
df['predictions']=predictions
print(df)
```

```python
from sklearn.metrics import r2_score
print("R2:",r2_score(df.Calories, df.predictions))
```

```python
from sklearn import metrics
import numpy as np

print("MSE:",metrics.mean_squared_error(df.Calories,df.predictions))
print("RMSE:",np.sqrt(metrics.mean_squared_error(df.Calories,df.predictions)))
```
